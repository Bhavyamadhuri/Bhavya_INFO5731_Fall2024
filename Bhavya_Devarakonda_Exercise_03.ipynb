## Question 2 (10 Points)
Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction.

```python
# Import necessary libraries
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import nltk
from nltk import pos_tag, word_tokenize
from nltk.corpus import sentiwordnet as swn
from nltk.util import ngrams
from collections import Counter
import numpy as np

# Download NLTK resources
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('sentiwordnet')
nltk.download('wordnet')

# tweet-like text data
sample_data = [
    "I love this product, it's amazing!",
    "The service was terrible and slow.",
    "I'm very happy with the results.",
    "Not great, not terrible, just okay.",
    "Absolutely fantastic experience!"
]

# Create a DataFrame
df = pd.DataFrame({'Tweet': sample_data})

# Feature 1: Bag of Words (BoW) using CountVectorizer
# This will count the frequency of each word in the text
bow_vectorizer = CountVectorizer()
bow_features = bow_vectorizer.fit_transform(df['Tweet'])

# Display the BoW features as a DataFrame for easy understanding
bow_df = pd.DataFrame(bow_features.toarray(), columns=bow_vectorizer.get_feature_names_out())
print("Bag of Words Features:\n", bow_df)

# Feature 2: TF-IDF (Term Frequency-Inverse Document Frequency)
# This will weigh the importance of each word based on how unique it is across the dataset
tfidf_vectorizer = TfidfVectorizer()
tfidf_features = tfidf_vectorizer.fit_transform(df['Tweet'])

# Display the TF-IDF features as a DataFrame
tfidf_df = pd.DataFrame(tfidf_features.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
print("\nTF-IDF Features:\n", tfidf_df)

# Feature 3: Sentiment Lexicon using SentiWordNet
# We'll calculate a sentiment score for each word in the text based on SentiWordNet

def get_wordnet_pos(tag):
    """Converts POS tag from nltk format to WordNet format."""
    if tag.startswith('J'):
        return 'a'  # Adjective
    elif tag.startswith('V'):
        return 'v'  # Verb
    elif tag.startswith('N'):
        return 'n'  # Noun
    elif tag.startswith('R'):
        return 'r'  # Adverb
    else:
        return None

def sentiment_score(text):
    """Calculates sentiment score using SentiWordNet for each word in the text."""
    tokens = word_tokenize(text)
    pos_tags = pos_tag(tokens)
    score = 0
    for word, tag in pos_tags:
        wn_tag = get_wordnet_pos(tag)
        if wn_tag:  # If the word has a valid POS tag
            synsets = list(swn.senti_synsets(word, wn_tag))
            if synsets:
                synset = synsets[0]  # Use the first synset (most common meaning)
                score += synset.pos_score() - synset.neg_score()  # Positive score - Negative score
    return score

# Apply sentiment_score function to each tweet
df['Sentiment_Score'] = df['Tweet'].apply(sentiment_score)
print("\nSentiment Lexicon Features:\n", df[['Tweet', 'Sentiment_Score']])

# Feature 4: Part-of-Speech (POS) Tagging
# We'll count the number of nouns, verbs, adjectives, and adverbs in each tweet

def pos_tag_count(text):
    """Counts the occurrence of POS tags in the text."""
    tokens = word_tokenize(text)
    pos_tags = pos_tag(tokens)
    pos_counts = Counter(tag for word, tag in pos_tags)
    return pos_counts

# Apply POS tag count function to each tweet
df['POS_Tags'] = df['Tweet'].apply(pos_tag_count)
print("\nPOS Tagging Features:\n", df[['Tweet', 'POS_Tags']])

# Feature 5: N-grams (Bigrams)
# We will extract bigrams (two-word sequences) from each tweet

def extract_bigrams(text):
    """Extracts bigrams (two-word sequences) from the text."""
    tokens = word_tokenize(text)
    bigrams = list(ngrams(tokens, 2))
    return bigrams

# Apply bigram extraction to each tweet
df['Bigrams'] = df['Tweet'].apply(extract_bigrams)
print("\nN-grams (Bigrams) Features:\n", df[['Tweet', 'Bigrams']])

```

```python
# You code here (Please add comments in the code):
# Import necessary libraries for feature selection
from sklearn.feature_selection import mutual_info_classif
from sklearn.preprocessing import LabelEncoder

# sentiment for each tweet:
# 1 = Positive, 0 = Negative
target = [1, 0, 1, 0, 1]

# Combine all extracted features into a single DataFrame
combined_features = pd.concat([bow_df, df[['Sentiment_Score']]], axis=1)

# Encode the target labels
le = LabelEncoder()
target_encoded = le.fit_transform(target)

# Apply Mutual Information to rank the features based on importance
mi_scores = mutual_info_classif(combined_features, target_encoded)

# feature importance scores to a data frame
mi_df = pd.DataFrame({
    'Feature': combined_features.columns,
    'Importance': mi_scores
})

# Sort the features by importance in descending order
mi_df_sorted = mi_df.sort_values(by='Importance', ascending=False)
print("\nFeature Importance Ranking (Mutual Information):\n", mi_df_sorted)

```

## Question 4 (10 points):
Write python code to rank the text based on text similarity. Based on the text data you used for question 2, design a query to match the most relevant docments. Please use the BERT model to represent both your query and the text data, then calculate the cosine similarity between the query and each text in your data. Rank the similary with descending order.

```python
import torch
from transformers import BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load pre-trained BERT model and tokenizer
bert_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = BertModel.from_pretrained('bert-base-uncased')

# Sample text data
texts = [
    "I love this product, it's amazing!",
    "The service was terrible and slow.",
    "I'm very happy with the results.",
    "Not great, not terrible, just okay.",
    "Absolutely fantastic experience!"
]

# Query text to find similar documents
search_query = "I love the new phone, it's amazing!"

# Get BERT embeddings for a given text
def generate_bert_embedding(text):
    encoded_input = bert_tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=128)
    with torch.no_grad():
        model_output = bert_model(**encoded_input)
    cls_vector = model_output.last_hidden_state[:, 0, :].squeeze().numpy()
    return cls_vector

# Get embedding for the query
query_vector = generate_bert_embedding(search_query)

# Get embeddings for all texts
text_vectors = [generate_bert_embedding(text) for text in texts]

# Compute cosine similarity between the query and each text
similarity_scores = [cosine_similarity([query_vector], [text_vector])[0][0] for text_vector in text_vectors]

# Combine the texts and their similarity scores
results = list(zip(texts, similarity_scores))

# Sort the results by similarity in descending order
sorted_results = sorted(results, key=lambda x: x[1], reverse=True)

# Display the ranked results
print("Ranked texts based on similarity to the query:\n")
for rank, (text, score) in enumerate(sorted_results, 1):
    print(f"{rank}. Similarity: {score:.4f} | Text: {text}")
```
